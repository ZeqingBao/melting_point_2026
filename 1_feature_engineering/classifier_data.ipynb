```python
import pandas as pd
import os

# 1. Load the data (already in your script)
df_combined = pd.read_parquet("../0_data/processed_data/data_with_selected_features_LGB_All_scaled.parquet")

# 2. Create the 'Binary' column using a mapping dictionary
# This converts 'L' -> 0 and 'H' -> 1
df_combined['Binary'] = df_combined['MP_label'].map({'L': 0, 'H': 1})

# 3. Define the directory and filename
data_prefix = '../0_data/processed_data/'
output_filename = 'data_with_selected_features_LGB_All_scaled_binary.parquet'
output_path = os.path.join(data_prefix, output_filename)

# 4. Save the file
# Ensure the directory exists first
os.makedirs(data_prefix, exist_ok=True)
df_combined.to_parquet(output_path, index=False)

print(f"File saved successfully to: {output_path}")
print(df_combined[['MP_label', 'Binary']].head())
```
