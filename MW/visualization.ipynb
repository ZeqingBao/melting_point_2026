```python
import matplotlib.pyplot as plt
import pandas as pd

# ==========================
# 1️⃣  EDIT YOUR DATA HERE
# ==========================

df = pd.DataFrame({
    "Model": [
        "Low + High MP",
        "Low MP",
        "High MP",
        "Best TL (no freeze)",
        "TL (freeze 1)",
        "TL (freeze 2)",
    ],
    "LowMP RMSE":  [59.998,34.895,190.556, 175.612, 243.798, 255.505],
    "HighMP RMSE": [97.249, 99.458, 28.283, 26.943, 30.483, 36.624],
    "Total RMSE":  [61.8611, 39.4191, 186.910, 172.257, 239.102, 250.608],
})

# ==========================
# 2️⃣  STYLE SETTINGS
# ==========================

FONT_FAMILY = "Helvetica"   # Try: "Arial", "Helvetica"
TITLE_SIZE = 20
TABLE_FONT_SIZE = 16

HEADER_COLOR = "#EAEAF2"
TL_ROW_COLOR = "#F7FBFF"

# ==========================
# 3️⃣  CREATE TABLE
# ==========================

plt.rcParams["font.family"] = FONT_FAMILY

fig, ax = plt.subplots(figsize=(10, 3))
ax.axis("off")

table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc="center",
    loc="center"
)

table.auto_set_font_size(False)
table.set_fontsize(TABLE_FONT_SIZE)
table.scale(1.2, 1.8)

# ==========================
# 4️⃣  STYLE CELLS
# ==========================

for (row, col), cell in table.get_celld().items():
    if row == 0:  # Header row
        cell.set_text_props(weight='bold')
        cell.set_facecolor(HEADER_COLOR)
    else:
        # Light highlight for transfer learning rows
        if "TL" in df.iloc[row-1, 0]:
            cell.set_facecolor(TL_ROW_COLOR)

# Bold best HighMW value
best_highmp_row = df["HighMP RMSE"].idxmin() + 1
table[(best_highmp_row, 2)].set_text_props(weight="bold")

plt.title("Model Performance Summary", fontsize=TITLE_SIZE, pad=20)

plt.tight_layout()

# ==========================
# 5️⃣  SAVE (Poster Ready)
# ==========================

#plt.savefig("poster_table.png", dpi=400, bbox_inches="tight")
#plt.savefig("poster_table.pdf", bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --------------------
# Data (from your table)
# --------------------
df = pd.DataFrame({
    "Model": [
        "Low",
        "High",
        "High_TL_no_freeze",
        "High_TL_fc1",
        "High_TL_fc1_fc2",
    ],
    "Test_LowMW": [36.093, 45.368, 41.968, 42.036, 44.445],
    "Test_HighMW": [48.014, 40.910, 40.671, 40.697, 44.426],
})

# Optional: if you want to sort by Total RMSE (add totals then sort)
# df["Total"] = [39.580, 44.238, 41.630, 41.688, 44.440]
# df = df.sort_values("Total", ascending=True).reset_index(drop=True)

# --------------------
# Plot: grouped bar chart + value labels
# --------------------
x = np.arange(len(df))
width = 0.38

plt.figure(figsize=(10, 5))
bars1 = plt.bar(x - width/2, df["Test_LowMW"], width, label="Test (LowMW)")
bars2 = plt.bar(x + width/2, df["Test_HighMW"], width, label="Test (HighMW)")

plt.xticks(x, df["Model"], rotation=30, ha="right")
plt.ylabel("RMSE")
plt.title("Model RMSE by Molecular-Weight Group (threshold = 300)")
plt.legend(frameon=False)

# Add value labels on top of bars
def add_labels(bars, fmt="{:.3f}", ypad=0.3):
    for b in bars:
        h = b.get_height()
        plt.text(
            b.get_x() + b.get_width()/2,
            h + ypad,
            fmt.format(h),
            ha="center",
            va="bottom",
            fontsize=9
        )

add_labels(bars1)
add_labels(bars2)

plt.tight_layout()
plt.show()

# --------------------
# Optional: save high-res figure
# --------------------
# plt.savefig("rmse_by_mw_group.png", dpi=300, bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
df = pd.DataFrame({
    "Model": [
        "High",
        "High_TL_no_freeze",
        "High_TL_fc1",
        "High_TL_fc1_fc2",
    ],
    "HighMW_RMSE": [40.910, 40.671, 40.697, 44.426]
})

# Baseline = High
baseline = df.loc[df["Model"] == "High", "HighMW_RMSE"].values[0]
df["Delta_RMSE"] = df["HighMW_RMSE"] - baseline

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(df["Model"], df["Delta_RMSE"])

plt.axhline(0, linestyle="--")
plt.ylabel("Δ RMSE vs Baseline (High)")
plt.title("Transfer Learning Improvement on HighMW")

plt.xticks(rotation=25, ha="right")

for b in bars:
    h = b.get_height()
    plt.text(b.get_x() + b.get_width()/2, h,
             f"{h:.3f}", ha="center",
             va="bottom" if h>=0 else "top")

plt.tight_layout()
plt.show()
```
