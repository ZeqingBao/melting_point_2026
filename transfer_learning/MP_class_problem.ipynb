```python
from pathlib import Path

BASE = Path.cwd()  
train_df = pd.read_csv("artifacts/final_train_df_scaled.csv")
test_df = pd.read_csv("artifacts/final_test_df_with_rdkit_scaled.csv")
```

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    balanced_accuracy_score,
    precision_recall_fscore_support,
    roc_auc_score,
)


GATE_FEATURES = [
    "BertzCT",
    "RingCount",
    "HallKierAlpha",
    "TPSA",
    "NOCount",
    "AvgIpc",
]

MP_CUTOFF_T = 200.0



SIGMOID_S = 20.0  # typical 10–30

def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1.0 / (1.0 + np.exp(-x))

def evaluate_router_on_testset(
    train_df: pd.DataFrame,
    test_df: pd.DataFrame,
    mp_cutoff: float,
    gate_features: list[str],
    train_target_col: str = "MP",
    test_target_col: str = "MP",
    use_ridge: bool = True,
    ridge_alpha: float = 1.0,
    threshold: float | None = None,
):
    """
    Fit router on train_df (using train_target_col) and evaluate on test_df (using test_target_col).
    """

    # ---------- TRAIN ----------
    train_data = train_df.dropna(subset=gate_features + [train_target_col]).copy()
    X_train = train_data[gate_features].values.astype(float)
    y_mp_train = train_data[train_target_col].values.astype(float)

    # ---------- TEST ----------
    # Need gate features at minimum
    test_data = test_df.dropna(subset=gate_features).copy()
    X_test = test_data[gate_features].values.astype(float)

    # Do we have true MP in test?
    test_has_mp = (test_target_col in test_data.columns) and test_data[test_target_col].notna().all()

    if test_has_mp:
        y_mp_test = test_data[test_target_col].values.astype(float)
        y_true_test = (y_mp_test >= mp_cutoff).astype(int)
    else:
        y_mp_test = None
        y_true_test = None

    # ---------- Scale + fit gate ----------
    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_test_s = scaler.transform(X_test)

    gate_model = Ridge(alpha=ridge_alpha, random_state=42) if use_ridge else LinearRegression()
    gate_model.fit(X_train_s, y_mp_train)

    # ---------- Predict gate MP on test ----------
    mp_gate_pred_test = gate_model.predict(X_test_s)

    # Router threshold on gate output
    if threshold is None:
        threshold = mp_cutoff

    p_high_test = sigmoid((mp_gate_pred_test - threshold) / SIGMOID_S)
    y_pred_test = (mp_gate_pred_test >= threshold).astype(int)

    # ---------- Metrics ----------
    if test_has_mp:
        cm = confusion_matrix(y_true_test, y_pred_test, labels=[0, 1])
        acc = accuracy_score(y_true_test, y_pred_test)
        bal_acc = balanced_accuracy_score(y_true_test, y_pred_test)
        prec, rec, f1, _ = precision_recall_fscore_support(
            y_true_test, y_pred_test, average="binary", zero_division=0
        )
        auc = roc_auc_score(y_true_test, p_high_test) if len(np.unique(y_true_test)) == 2 else np.nan

        print("\n=== ROUTER-ONLY (TRAIN -> TEST) EVALUATION ===")
        print(f"Train target: {train_target_col} | Test target: {test_target_col}")
        print(f"MP cutoff (true labels): {mp_cutoff}")
        print(f"Gate threshold used:     {threshold:.2f}")
        print("\nConfusion matrix (rows=true [low,high], cols=pred [low,high]):")
        print(cm)

        tn, fp, fn, tp = cm.ravel()
        print(f"\nTN={tn}  FP={fp}  FN={fn}  TP={tp}")
        print(f"Accuracy:          {acc:.3f}")
        print(f"Balanced accuracy: {bal_acc:.3f}")
        print(f"Precision (high):  {prec:.3f}")
        print(f"Recall (high):     {rec:.3f}")
        print(f"F1 (high):         {f1:.3f}")
        print(f"ROC-AUC (soft):    {auc:.3f}")
    else:
        cm = None
        print("\n[Test set has no complete MP column → returning predictions only, no metrics.]")

    # ---------- Coefficients ----------
    print("\nGate coefficients (on standardized features):")
    for name, coef in zip(gate_features, gate_model.coef_):
        print(f"  {name:15s}: {coef:+.4f}")
    print(f"Intercept: {gate_model.intercept_:+.4f}")

    # ---------- Attach predictions ----------
    out = test_data.copy()
    out["mp_gate_pred"] = mp_gate_pred_test
    out["p_high"] = p_high_test
    out["router_pred_label"] = y_pred_test
    if test_has_mp:
        out["router_true_label"] = y_true_test

    return {
        "scaler": scaler,
        "gate_model": gate_model,
        "threshold": threshold,
        "pred_table": out,
        "cm": cm,
    }
```

```python
result = evaluate_router_on_testset(
    train_df=train_df,
    test_df=test_df,
    mp_cutoff=250.0,  # your cutoff
    gate_features=["BertzCT","RingCount", "HallKierAlpha","TPSA", "NOCount","AvgIpc"],
    train_target_col="MP",
    test_target_col="exp MP",
    use_ridge=True,
    ridge_alpha=1.0,
)
```
