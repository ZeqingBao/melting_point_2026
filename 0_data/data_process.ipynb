```python
import os
import pandas as pd
```

```python
def combine_data(folder_name, processed=0):

    full_input_path = os.path.join("original_curated/",folder_name)
    full_output_path = os.path.join("processed_data/", folder_name)

    if processed:
        full_input_path = folder_name

    if not os.path.exists(full_input_path):
        raise FileNotFoundError(f"Folder not found: {full_input_path}")


    print(f"Processing folder: {full_input_path}")
    print()
    # Step 1: List all CSV files
    csv_files = [f for f in os.listdir(full_input_path) if f.endswith('.csv')]
    print("CSV files found:")
    for f in csv_files:
        print(f)

    if not csv_files:
        raise ValueError(f"No CSV files found in {full_input_path}")

    # Step 2: Combine all CSV files
    df_list = []
    for f in csv_files:
        file_path = os.path.join(full_input_path, f)
        df = pd.read_csv(file_path)
        df_list.append(df)

    combined_df = pd.concat(df_list, ignore_index=True)
    
    if not processed:
        combined_df["Type"] = folder_name
    print()
    print(f"Number of rows after combination: {len(combined_df)}")

    # Step 3: Remove duplicates based on 'SMILES'
    if 'SMILES' not in combined_df.columns:
        raise KeyError("The column 'SMILES' is not present in the combined dataframe.")
    
    deduped_df = combined_df.drop_duplicates(subset='SMILES')
    print(f"Number of rows after removing duplicates based on 'SIMLES': {len(deduped_df)}")

    # Step 4: Reset index
    deduped_df.reset_index(drop=True, inplace=True)

    # Step 5: Save to CSV if output_path is provided

    deduped_df.to_csv(full_output_path + ".csv", index=False)
    print(f"Combined data saved to {full_output_path}.csv")

    return deduped_df
```

```python
combined_train_DA = combine_data('train_with_data_augmentation')
combined_train_DA.tail(20)
```
