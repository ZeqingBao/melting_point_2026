```python
#Importing all the relevant libraries
import os
import pandas as pd
from pathlib import Path
from collections import defaultdict 
```

```python
#Set all the file paths
#processed_data_embeddings = file that Nasim sent us 
#original curated = duplicate of folder structure of original datasets
#target root = where the original datasets + embeddings coloumn will be exported 

# 1) Define the base folder *relative* to your notebook/script
BASE = Path.cwd().parent
print(BASE)

# 2) Import the test file
test_file_dir = BASE / "data_curation" / "original_curated" / "test_predictions" / "consensus_without_data_augmentation.csv"

# 3) Load dataframe 
test_df = pd.read_csv(test_file_dir)
test_df.info()
print(test_df.isna().sum())

master_path = BASE / "data_curation" / "import_embeddings" / "processed_data_embeddings.csv"
source_root = BASE / "data_curation" / "original_curated"
target_root = BASE / "data_curation" / "original_curated_with_embeddings_and_MW"
```

```python
#Load the processed embeddings file
master_df = pd.read_csv(master_path)
# Zip maps SMILES with embedding, and dict turns it into dictionary
embedding_dict = dict(zip(master_df["SMILES"], master_df["embeddings"]))
#Check the length of the dictionary to see if they all were imported 
print(len(embedding_dict)) #Python can do this by itself, instead of printing and the user checking if the rows on the excel file match the printed number, just tell python to see if the 2 numbers match, if True continute
#Check the first 10 items to see if they are accurate, and see if any SMILES were duplicated
duplicate_count = master_df["SMILES"].duplicated().sum()
print(duplicate_count)
print(master_df["embeddings"].isna().sum())
list(embedding_dict.items())[:5]
```

```python
master_df.info()
#Identify which rows are missing embeddings
missing_embeddings = master_df["embeddings"].isna()
#Creates a new df
missing_embeddings = master_df.loc[missing_embeddings]
missing_embeddings.to_csv('missing_embeddings.csv', index=False)
```

```python
#Creates a dictionary (even if the key does not have to exist yet)
unmatched_counts = defaultdict(list)
subfolders = ["test_predictions", "train_with_data_augmentation", "train_without_data_augmentation", "val"]

```

```python
#Loops through the folders in the test, training (both with and without augmentation) and val folders
for subfolder in subfolders:
    source_folder = source_root / subfolder
    #Builds a path to save the updated files
    target_folder = target_root / subfolder
    target_folder.mkdir(parents=True, exist_ok=True)  # Create output folders if they don't exist

    # Loop through all CSV files in the folder in original_curated, .glob matches to the .csv files
    for file in source_folder.glob("*.csv"):
        df = pd.read_csv(file)

        # Creates a new embeddings coloumn, looks at each SMILES coloumn and uses embedding_dict to match the corresponding embedding
        #Nan for missing values
        df["embeddings"] = df["SMILES"].map(embedding_dict)

        # Count how many SMILES did not have a matching embedding
        unmatched = df["embeddings"].isna().sum()
        unmatched_counts[subfolder].append((file.name, unmatched))

        # Save the updated DataFrame to the new location
        df.to_csv(target_folder / file.name, index=False)
```

```python
#Create an empty list secreibing the folder, file and how many SMILES were unmatched
summary = []

for folder, files in unmatched_counts.items():
    for file_name, unmatched in files:
        summary.append({
            "Folder": folder,
            "File": file_name,
            "Unmatched_SMILES": unmatched
        })

unmatched_df = pd.DataFrame(summary)
unmatched_df.sort_values(by=["Folder", "Unmatched_SMILES"], ascending=[True, False], inplace=True)
unmatched_df.reset_index(drop=True, inplace=True)

# Display it nicely in a notebook
display(unmatched_df)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

def molecular_weight(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            return Descriptors.MolWt(mol)
        else:
            return None
    except:
        return None
```

```python
BASE = Path.cwd().parent
subfolders = ["test_predictions", "train_with_data_augmentation", "train_without_data_augmentation", "val"]
target_root = BASE / "data_curation" / "original_curated_with_embeddings_and_MW"

summary = []

#Loops through the folders in the test, training (both with and without augmentation) and val folders in embeddings folder
for subfolder in subfolders:
    folder = target_root / subfolder
    if not folder.exists():
        print(f"Folder {folder} does not exist. Skipping.")
        continue
    
    # Loop through all CSV files in the folder in original_curated, .glob matches to the .csv files
    for file in folder.glob("*.csv"):
        df = pd.read_csv(file)
        # Check if the SMILES column exists
        if "SMILES" not in df.columns:
            print(f"SMILES column not found in {file.name}. Skipping.")
            continue
    
        # Creates a new molecular weight column, looks at each SMILES column and calculates the molecular weight
        df["MW"] = df["SMILES"].apply(molecular_weight).astype("Float64").round(3)

        # Overwrite the existing file with the new DataFrame
        df.to_csv(file, index=False)
        
        df.info()
        rows = len(df)
        coloumn_names = df.columns.tolist()
        missing_SMILES = int(df['SMILES'].isna().sum())
        missing_MW = int(df['MW'].isna().sum())
        missing_embeddings = int(df['embeddings'].isna().sum())
        
        summary.append({
            "Folder": subfolder,
            "File": file.name,
            "Coloumn_Names": coloumn_names,
            "Total_Rows": rows,
            "Missing_SMILES": missing_SMILES,
            "Missing_MW": missing_MW,
            "Missing_Embeddings": missing_embeddings
        })

summary_df = (
    pd.DataFrame(summary)
      .sort_values(by=["Folder"])
      .reset_index(drop=True)
) 
display(summary_df)       
```
